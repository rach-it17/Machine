Import Python Libraries

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
```

Load Data Into DataFrame

```python
df=pd.read_csv("terror_info.csv",encoding='ISO-8859-1')
```

Selecting the Key Features From the DataSet

```python
df=df[['iyear','imonth','country','region_txt','provstate','city','latitude','longitude','summary','success',
       'attacktype1_txt','attacktype2_txt','attacktype3_txt',
       'targtype1_txt','gname','motive','weaptype1_txt','nkill','nwound',]]
```

```python
df.isnull().sum()
```

```python
df.info()
```

Analysing the Data

```python
print("Country with the most attacks:",df['country'].value_counts().idxmax())
print("Most Attack Types:",df['attacktype1_txt'].value_counts().idxmax())
print("Group with the most attacks:",df['gname'].value_counts().index[1])
print("Region with the most attacks:",df['region_txt'].value_counts().idxmax())
print("Year with most attacks:",df['iyear'].value_counts().idxmax())
print("Month with the most attacks:",df['imonth'].value_counts().idxmax())

```

Number of Attacks Per Year

```python
df['iyear'].value_counts(dropna = False).sort_index()
```

Data visualization

```python
x_year = df['iyear'].unique()
y_count_years = df['iyear'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sn.barplot(x = x_year,
           y = y_count_years,
           palette = 'mako')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

```python
pd.crosstab(df.iyear, df.region_txt).plot(kind='area',figsize=(20,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```
